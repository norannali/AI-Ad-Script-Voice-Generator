## Step 3: Initialize the LLM

```python
from langchain_community.llms import Ollama
llm = Ollama(
    model="ollama/deepseek-r1:1.5b",
    base_url="http://localhost:11434"
)
```

##  Step 4‚Äì6: Enter Ad Details, Generate Prompt, and Get the Script
This section will:
1. Prompt you to enter the ad details (product, audience, platform, duration),
2. Automatically generate a prompt for the LLM,
3. Call the DeepSeek model via Ollama and return a natural-sounding English ad script.


```python
from ollama import Client

# Connect to the local Ollama server
client = Client()

# ==== Step 1: Get user inputs ====
product = input("üîπ Enter your product and a short description (e.g., Smartwatch that tracks heart rate, steps, and sleep):\n> ")
audience = input("üîπ Who is your target audience? (e.g., young adults into fitness and tech):\n> ")
platform = input("üîπ Which platform will you use this ad for? (e.g., TikTok - Instagram):\n> ")
duration = input("üîπ What's the ad duration in seconds? (e.g., 30):\n> ")

# ==== Step 2: Create the prompt ====
prompt = f"""
You are a professional social media ad copywriter.

Generate a short, punchy, and exciting ad script in **English only** for a {duration}-second video.

üéØ Product:
{product}

üì± Platform:
{platform}

üë• Target Audience:
{audience}

üó£ Tone:
Energetic, modern, emotionally engaging, and ends with a strong call to action.

üé¨ Output Format:

English Version:
<Your English ad script here>

Make it sound natural and realistic, like a TikTok creator would say it in a real video.
"""

# ==== Step 3: Call the model ====
response = client.chat(
    model="deepseek-r1:1.5b",
    messages=[{"role": "user", "content": prompt}]
)

# ==== Step 4: Extract and store the ad script ====
full_response = response['message']['content']


ad_script = full_response.split('</think>')[-1].strip()

# ==== Step 5: Print results ====
print("\nüéØ Generated English Ad Script:\n")
print(ad_script)
```

##  Step 7: Convert Generated Ad Script to Speech (TTS)

This step uses `edge-tts` to convert the generated ad script into natural-sounding speech using Microsoft's `Aria` voice model.

- ‚úÖ Removes symbols, emojis, and unnecessary tags
- ‚úÖ Saves as `output_voice.mp3`
- ‚úÖ Plays the audio directly in Colab


```python
import edge_tts
import asyncio
import re

# Text to be converted to speech (without symbols or emojis)
text = ad_script

# Function to clean text from symbols and emojis
def clean_text(text):
    # Remove bold text (content between **)
    #text = re.sub(r'\*\*.*?\*\*', '', text)
    # Remove any content between <> tags (like <think>)
    text = re.sub(r'<.*?>', '', text)
    # Remove hashtags
    text = re.sub(r'#\w+', '', text)
    # Remove emojis and special symbols (keeping basic punctuation)
    text = re.sub(r'[^\w\s.,!?]', '', text)
    # Remove multiple empty lines
    text = re.sub(r'\n\s*\n', '\n', text)
    return text.strip()

# Clean the text
clean_text = clean_text(text)

# Output audio file path
output_path = "output_voice.mp3"

# Function to generate speech
async def generate_speech():
    communicate = edge_tts.Communicate(clean_text, voice="en-US-AriaNeural")
    await communicate.save(output_path)

# Run the function
await generate_speech()

# Play the generated audio
from IPython.display import Audio
Audio("output_voice.mp3")
```

## Step 8 :   AI-Powered Ad Script & Voice Generator Web App

This interactive Gradio app lets you:

‚úÖ Generate a **short and punchy English ad script** for a product  
‚úÖ Automatically convert the script into **natural-sounding speech** using Microsoft Edge TTS  
‚úÖ Supports **TikTok, Instagram Reels, YouTube Shorts**, and more  
‚úÖ Powered by **Ollama (DeepSeek 1.5B)** and **Edge-TTS**


```python
import gradio as gr
from ollama import Client
import asyncio
import edge_tts
import tempfile
import os
import re
from pydub import AudioSegment

# Connect to local Ollama server
client = Client(host='http://localhost:11434')

# Clean text to be safe for TTS
def clean_text(text):
    text = re.sub(r'<.*?>', '', text)
    text = re.sub(r'#\w+', '', text)
    text = re.sub(r'[^\w\s.,!?]', '', text)
    text = re.sub(r'\n\s*\n', '\n', text)
    return text.strip()

# Split long text into shorter chunks for better TTS duration accuracy
def split_text(text, max_chars=350):
    sentences = re.split(r'(?<=[.!?])\s+', text)
    chunks = []
    current = ''
    for sentence in sentences:
        if len(current) + len(sentence) < max_chars:
            current += ' ' + sentence
        else:
            chunks.append(current.strip())
            current = sentence
    if current:
        chunks.append(current.strip())
    return chunks

# Async TTS for multiple chunks
async def tts_async_chunks(text_chunks, voice="en-US-AriaNeural"):
    audio_files = []
    for i, chunk in enumerate(text_chunks):
        tmp_audio = tempfile.NamedTemporaryFile(delete=False, suffix=".mp3")
        audio_files.append(tmp_audio.name)
        communicate = edge_tts.Communicate(chunk, voice=voice)
        await communicate.save(tmp_audio.name)
    return audio_files

# Merge multiple mp3 files into one
def merge_audio(audio_paths, output_path):
    combined = AudioSegment.empty()
    for path in audio_paths:
        combined += AudioSegment.from_file(path, format="mp3")
    combined.export(output_path, format="mp3")

# Main function: Prompt ‚Üí Generate Script ‚Üí Clean ‚Üí TTS ‚Üí Output
def generate_ad(product, audience, platform, duration):
    approx_words = int(duration * 2.85)
    prompt = f"""
You are a professional social media ad copywriter.

Generate a short, punchy, and exciting ad script in **English only** that fits within approximately {duration} seconds (about {approx_words} words).

üéØ Product:
{product}

üì± Platform:
{platform}

üë• Target Audience:
{audience}

üó£ Tone:
Energetic, modern, emotionally engaging, and ends with a strong call to action.

üé¨ Output Format:

English Version:
<Your English ad script here>

Make it sound natural and realistic, like a TikTok creator would say it in a real video.
    """

    try:
        response = client.chat(
            model="deepseek-r1:1.5b",
            messages=[{"role": "user", "content": prompt}]
        )
        full_response = response['message']['content']
    except Exception as e:
        return f"Error: {e}", None

    ad_script = full_response.split('</think>')[-1].strip()
    cleaned = clean_text(ad_script)
    text_chunks = split_text(cleaned)

    # Create temp audio output file
    with tempfile.NamedTemporaryFile(delete=False, suffix=".mp3") as tmpfile:
        final_audio_path = tmpfile.name

    # Generate audio and merge
    asyncio.run(_generate_and_merge_audio(text_chunks, final_audio_path))

    return cleaned, final_audio_path

# Wrapper for async audio generation + merging
async def _generate_and_merge_audio(text_chunks, final_audio_path):
    audio_paths = await tts_async_chunks(text_chunks)
    merge_audio(audio_paths, final_audio_path)
    # Clean up temp chunk files
    for f in audio_paths:
        if os.path.exists(f):
            os.remove(f)

# Gradio Interface
with gr.Blocks() as demo:
    gr.Markdown("# üé¨ AI Ad Script + Voice Generator")
    gr.Markdown("Generate realistic social media video ads with voiceovers using Ollama & Edge TTS.")

    with gr.Row():
        product = gr.Textbox(label="üîπ Product Description", placeholder="e.g., Smartwatch that tracks steps, heart rate...")
        audience = gr.Textbox(label="üë• Target Audience", placeholder="e.g., Tech-savvy fitness enthusiasts")
    with gr.Row():
        platform = gr.Dropdown(choices=["TikTok", "Instagram", "YouTube Shorts"], label="üì± Platform", value="TikTok")
        duration = gr.Slider(minimum=15, maximum=60, value=30, label="‚è±Ô∏è Ad Duration (seconds)")
    submit_btn = gr.Button("üéØ Generate Ad")

    output_text = gr.Textbox(label="üì¢ Generated Script")
    output_audio = gr.Audio(label="üîä Voice Output", type="filepath")

    submit_btn.click(fn=generate_ad, inputs=[product, audience, platform, duration], outputs=[output_text, output_audio])

demo.launch()
```
